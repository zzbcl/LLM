```python
from transformers import AutoModelForCausalLM, BitsAndBytesConfig
from deepseek_vl2.models import DeepseekVLV2Processor
import torch
import re
# === 模型路径（本地） ===
model_path = "/content/gdrive/MyDrive/Notebooks/Hashing/deepseek/deepseek-vl2-tiny"

# === 4bit 量化配置 ===
bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_use_double_quant=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype=torch.float16
)

# === 加载模型 + tokenizer ===
processor = DeepseekVLV2Processor.from_pretrained(model_path)
tokenizer = processor.tokenizer

# ❌ 暂时不用量化（先跑通功能）
model = AutoModelForCausalLM.from_pretrained(
    model_path,
    trust_remote_code=True,
    device_map="auto"
)


conversation = [
    {
        "role": "<|User|>",
        "content": (
            "请判断以下新闻是否真实（0=假新闻，1=真新闻），请只返回一个数字（0或1），不要多余内容，直接输出：\n"
            "标题：The Pope Flat Out Called Fox News Type Journalism ‘Terrorism’\n"
            "内容：Fox News 的报道激发了人们的恐惧，教皇对此表示不满……"
        ),
    },
    {"role": "<|Assistant|>", "content": ""}
]


# === 准备模型输入（不传 images） ===
inputs = processor(
    conversations=conversation,
    images=[],  # ✅ 显式传入空图像列表
    force_batchify=True,
    system_prompt=""
)

# === 推理 ===
device = model.device  # 获取模型所在设备

outputs = model.generate(
    input_ids=inputs.input_ids.to(device),              # ✅ 移到 GPU
    attention_mask=inputs.attention_mask.to(device),    # ✅ 同上
    pad_token_id=tokenizer.eos_token_id,
    max_new_tokens=1,
    do_sample=False
)



# 解析输出
result = tokenizer.decode(outputs[0], skip_special_tokens=True)
print("🧠 原始输出：", result)



# 获取最后一个数字 0 或 1（非 greedy，取最后一个匹配）
matches = re.findall(r"[01]", result)
label = matches[-1] if matches else "❓未识别"

print("✅ 最后一个数字作为预测标签：", label)
```

```python
# 微调
import os
from transformers import (
    AutoModelForCausalLM, TrainingArguments, Trainer, BitsAndBytesConfig
)
from datasets import load_dataset
from peft import LoraConfig, get_peft_model
from deepseek_vl2.models import DeepseekVLV2Processor
import torch

# === [步骤 1] 设置路径与参数 ===
model_path = "/content/gdrive/MyDrive/Notebooks/Hashing/deepseek/deepseek-vl2-tiny"
data_path = "/content/gdrive/MyDrive/Notebooks/Hashing/deepseek/fake_news/train_deepseek.jsonl"
output_dir = "/content/gdrive/MyDrive/Notebooks/Hashing/deepseek/Zelular_outputs/deepseek-sft-optimized"

# === [步骤 2] 设置 4-bit 量化配置 ===
bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_use_double_quant=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype=torch.float16
)

# === [步骤 3] 加载模型与 Tokenizer ===
processor = DeepseekVLV2Processor.from_pretrained(model_path)
tokenizer = processor.tokenizer

model = AutoModelForCausalLM.from_pretrained(
    model_path,
    trust_remote_code=True
)
model.config.use_cache = False  # 必须禁用 use_cache 否则 Trainer 会报错

# === [步骤 4] 应用 LoRA 配置（优化 LoRA 参数） ===
peft_config = LoraConfig(
    r=8,                          # 增大 r
    lora_alpha=16,                 # 增大 lora_alpha
    target_modules=["q_proj", "v_proj"],
    lora_dropout=0.1,              # 更高的 dropout
    bias="none",
    task_type="CAUSAL_LM"
)
model = get_peft_model(model, peft_config)

# === [步骤 5] 加载训练数据集 ===
dataset = load_dataset("json", data_files=data_path, split="train")

# === [步骤 6] Tokenize 样本并对齐 Labels ===
def preprocess(example):
    prompt = example["text"] + "\n答案："
    target = str(example["label"])

    full_input = prompt + target
    tokenized = tokenizer(full_input, truncation=True, max_length=512, padding="max_length")
    input_ids = tokenized["input_ids"]
    label_token_ids = tokenizer(target, add_special_tokens=False)["input_ids"]

    # 用 -100 mask prompt 部分
    labels = [-100] * (len(input_ids) - len(label_token_ids)) + label_token_ids
    labels = labels[:512] + [-100] * max(0, 512 - len(labels))  # 确保长度对齐

    return {
        "input_ids": input_ids,
        "attention_mask": tokenized["attention_mask"],
        "labels": labels
    }

dataset = dataset.map(preprocess, remove_columns=dataset.column_names)

# === [步骤 7] 设置训练参数 ===
training_args = TrainingArguments(
    output_dir=output_dir,
    per_device_train_batch_size=1,     # 使用较小的 batch size
    gradient_accumulation_steps=2,     # 更低的累积步数
    num_train_epochs=3,                # 减少训练轮数
    learning_rate=2e-5,                # 降低学习率
    bf16=False,                        # 禁用 bfloat16
    fp16=False,                        # 禁用 float16
    logging_steps=10,
    save_steps=100,
    save_total_limit=2,
    report_to="none",
)

# === [步骤 8] 启动 Trainer 开始训练 ===
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=dataset,
    tokenizer=tokenizer
)

trainer.train()

# === [步骤 9] 保存微调模型 ===
model.save_pretrained(output_dir)
tokenizer.save_pretrained(output_dir)
print("✅ 微调完成并已保存至：", output_dir)
```

```python
# 模型评估
import os
import re
import torch
from datasets import load_dataset
from transformers import AutoModelForCausalLM, BitsAndBytesConfig
from peft import PeftModel
from deepseek_vl2.models import DeepseekVLV2Processor
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix

# === [1] 路径设定 ===
base_model_path = "/content/gdrive/MyDrive/Notebooks/Hashing/deepseek/deepseek-vl2-tiny"
finetuned_model_path = "/content/gdrive/MyDrive/Notebooks/Hashing/deepseek/Zelular_outputs/deepseek-sft-optimized"
test_path = "/content/gdrive/MyDrive/Notebooks/Hashing/deepseek/fake_news/test_deepseek.jsonl"

# === [2] 加载 Tokenizer 和 微调后模型 ===
processor = DeepseekVLV2Processor.from_pretrained(base_model_path)
tokenizer = processor.tokenizer

# 加载基座模型 + LoRA adapter
model = AutoModelForCausalLM.from_pretrained(
    base_model_path,
    trust_remote_code=True
)
model = PeftModel.from_pretrained(model, finetuned_model_path)
model = model.eval().cuda()  # 放到 GPU

# === [3] 加载测试数据 ===
dataset = load_dataset("json", data_files=test_path, split="train")

# === [4] 评估逻辑 ===
true_labels = []
pred_labels = []

print("🚀 开始测试集推理...")

for sample in dataset:
    text = sample["text"]
    label = str(sample["label"])
    true_labels.append(int(label))

    prompt = (
        "请判断以下新闻是否真实（0=假新闻，1=真新闻），请只返回一个数字（0或1），不要多余内容，直接输出：\n"
        + text + "\n答案："
    )
    conversation = [{"role": "<|User|>", "content": prompt}, {"role": "<|Assistant|>", "content": ""}]

    inputs = processor(
        conversations=conversation,
        images=[],
        force_batchify=True,
        system_prompt=""
    )
    input_ids = inputs.input_ids.to(model.device)
    attention_mask = inputs.attention_mask.to(model.device)

    with torch.no_grad():
        outputs = model.generate(
            input_ids=input_ids,
            attention_mask=attention_mask,
            pad_token_id=tokenizer.eos_token_id,
            max_new_tokens=1,
            do_sample=False
        )

    decoded = tokenizer.decode(outputs[0], skip_special_tokens=True)
    matches = re.findall(r"[01]", decoded)
    pred = int(matches[-1]) if matches else -1  # -1 = 无法解析
    pred_labels.append(pred)

    print(f"✅ 真值: {label} | 预测: {pred} | 原始输出: {decoded.strip()}")


# === [5] 计算多个评价指标 ===
valid_indices = [i for i, p in enumerate(pred_labels) if p in [0, 1]]
y_true = [true_labels[i] for i in valid_indices]
y_pred = [pred_labels[i] for i in valid_indices]

# 准确率
acc = accuracy_score(y_true, y_pred)

# 精确率
precision = precision_score(y_true, y_pred)

# 召回率
recall = recall_score(y_true, y_pred)

# F1分数
f1 = f1_score(y_true, y_pred)

# 混淆矩阵
conf_matrix = confusion_matrix(y_true, y_pred)

# 打印结果
print(f"\n🎯 评估完成！有效样本数: {len(valid_indices)} / {len(pred_labels)}")
print(f"📊 准确率：{acc:.4f}")
print(f"🔶 精确率：{precision:.4f}")
print(f"🔸 召回率：{recall:.4f}")
print(f"💡 F1 分数：{f1:.4f}")
print(f"📊 混淆矩阵：\n{conf_matrix}")
```
